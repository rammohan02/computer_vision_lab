```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("CED19I026_selfie.jpg")
# cv2.imshow("Original",img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)

ret, thres = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

titles = ['original', 'Gray', 'Binary']
images = [rgb, gray, thres]

for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.show()

print(f"Red Min={rgb[:,:,0].min()}\nRed Max={rgb[:,:,0].max()}\nGreen Min={rgb[:,:,1].min()}\nGreen Max={rgb[:,:,1].max()}\nBlue Min={rgb[:,:,2].min()}\nBlue Max={rgb[:,:,2].max()}\n")

imgc = cv2.merge((gray, gray, gray))
imgb = cv2.merge((thres, thres, thres))

final = np.hstack((img, imgc, imgb))
finalb = cv2.resize(final,(960, 540))
cv2.imshow("Final", finalb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
