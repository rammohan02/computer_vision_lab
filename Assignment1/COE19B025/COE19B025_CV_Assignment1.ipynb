```python
import cv2
```

```python
img = cv2.imread("COE19B025_selfie.jpg")
```

```python
cv2.imshow("Original", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(img)
```

```python
print(img.size)
```

```python
print(img.shape)
print(img[0,0])
print(img)
```

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb)
```

```python
gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray)
```

```python
plt.imshow(gray, 'gray')
```

```python
ret, thres = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(thres)
```

```python
cv2.imshow('binary', thres)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow('Gray', gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
titles = ['Original', 'Gray', 'binary']
images = [rgb, gray, thres]

for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
import numpy as np


imgc = cv2.merge((gray, gray, gray))
imgb = cv2.merge((thres, thres, thres))

final = np.hstack((img, imgc, imgb))
finalb = cv2.resize(final,(960, 540))
cv2.imshow("Final", finalb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
