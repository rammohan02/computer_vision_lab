```python
import cv2
from matplotlib import pyplot as plt
# tsukuba_l.png

img = cv2.imread("7.jpg", 0)
h, w = img.shape

img_l = img[:, :int(w/2)]
img_r = img[:, int(w/2):]

print(f"{img_l.shape} {img_r.shape}")

print(h, w)

# img1_l = cv2.imread("2_l.jpg", 0)
# img1_r = cv2.imread("2_r.jpg", 0)

# img1_l = cv2.resize(img1_l, (960, 540))
# img1_r = cv2.resize(img1_r, (960, 540))
# h, w = img1_r.shape
# h, w

stereo = cv2.StereoBM_create(numDisparities=32,  blockSize=21)
disparity = stereo.compute(img_l, img_r)

# disparity1 = cv2.convertScaleAbs(disparity)

plt.imshow(disparity, 'gray')
plt.show()
cv2.imshow("Img_L", img_l)
cv2.imshow("Img_R", img_r)
# cv2.imshow("Disparity", disparity)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
